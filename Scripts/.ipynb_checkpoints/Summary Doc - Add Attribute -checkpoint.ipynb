```python
import json
import pprint
```

```python
json_file_url = "../datasets/greater-hyderabad-municipal-corporation-budget-summary-statement.json"
```

```python
data = json.load(open(json_file_url))
```

```python
definition_dict = {"Revenue Expenditure (B2)" : "Expenditure not used to create assets e.g. Expenses on salaries of employees or other administrative costs. " ,
"Total Surplus/Deficit" : "The excess of total government expenditure over total receipts",
"Revenue Receipts (B1)" : "Mostly revenues from taxes, dividends from companies owned by the government, and user charges on some public services. ",
"Capital Receipts (C1)" : "Mainly funds borrowed by the government from various sources, both in India and from overseas, and repayments by state governments of loans from the centre",
"Capital Expenditure (C2)" : "Expenditure used to create assets or to reduce liabilities e.g.  building a road."}
```

```python
for index, obj in enumerate(data):
    try :
        definition_dict[obj["name"]]
        data[index]["def"] = definition_dict[obj["name"]]
    except :
        pass
```

```python
pp = pprint.PrettyPrinter(indent=4)
pp.pprint(data)
```

```python
with open(json_file_url, 'w') as outfile:
    json.dump(data, outfile)
```
